## **BREAKDOWN**

Trying to achieve the same result through steps:

Mean differences:

```python
def get_diff_mean(df_a: pd.DataFrame, df_b: pd.DataFrame, column: str):
    """
    Subtracts table `df_b` with `column` name from table `df_a` with column name `column`,
    and takes the average of the resulting column.

    This function will return a Series/array/list of differences in the same index
    of the two columns passed accessed with the `column` string argument in the `df_a`
    and `df_b` DataFrames.
    """
    return df_a[column].subtract(df_b[column]).mean()
```

```python
precision_diff_mean = get_diff_mean(a, b, "Precision")
recall_diff_mean = get_diff_mean(a, b, "Recall")
fscore_diff_mean = get_diff_mean(a, b, "F1-Score")
```

Standard deviations of each:

```python
def get_sd(df_a: pd.DataFrame, df_b: pd.DataFrame, column: str, mean):
    """
    Gets standard deviation of the differences from `df_a` and `df_b` with column name
    `column`, using the mean difference `mean`.
    """
    numerator = 0
    denominator = len(df_a[column]) - 1

    series = df_a[column].subtract(df_b[column])

    for value in series:
        numerator += (value - mean) ** 2
    
    return math.sqrt(numerator / denominator)
```

```python
precision_sd = get_sd(a, b, "Precision", precision_diff_mean)
recall_sd = get_sd(a, b, "Recall", recall_diff_mean)
f1score_sd = get_sd(a, b, "F1-Score", fscore_diff_mean) 
```

Finally, for t-test output:

```python
precision_ttest = precision_diff_mean / (precision_sd / math.sqrt(len(a["Precision"])))
precision_ttest
```

```python
recall_ttest = recall_diff_mean / (recall_sd / math.sqrt(len(a["Recall"])))
recall_ttest
```

```python
f1_score_ttest = fscore_diff_mean / (f1score_sd / math.sqrt(len(a["F1-Score"])))
f1_score_ttest
```

For p-value:

```python
# For one-tailed paired t-test, alpha level 0.05
# We're doing one-tailed since we are only concerned with significant changes in the positive direction.
# Source: https://www.sjsu.edu/faculty/gerstman/StatPrimer/t-table.pdf
critical_value = 1.697

precision_is_significant = precision_ttest > critical_value
recall_is_significant = recall_ttest > critical_value
f1score_is_significant = f1_score_ttest > critical_value

precision_is_significant, recall_is_significant, f1score_is_significant
```

Compile all values for uniformity:

```python
variables_df = pd.DataFrame({
    "Variable": ["n", "d", "Sd"],
    "Precision": [len(a["Precision"]), precision_diff_mean, precision_sd],
    "Recall": [len(a["Recall"]), recall_diff_mean, recall_sd],
    "F1-Score": [len(a["F1-Score"]), fscore_diff_mean, f1score_sd],
})

variables_df = variables_df.set_index("Variable")
variables_df.to_csv("variables.csv")
print("Done!")
```

```python
ttest_df = pd.DataFrame({
    "Metric": ["Precision", "Recall", "F1-Score"],
    "T-statistic": [precision_ttest, recall_ttest, f1_score_ttest],
    "Critical value": [ critical_value for _ in range(3) ],
    "Is significant": [precision_is_significant, recall_is_significant, f1score_is_significant]
})

ttest_df = ttest_df.set_index("Metric")
ttest_df.to_csv("ttest.csv")
print("Done!")
```
